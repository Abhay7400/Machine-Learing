```python
import pandas as pd
import numpy as np
```

```python
df=pd.read_csv('/content/titanic dummy (Mixed_data).csv')
```

```python
df.head()
```

```python
#A stand for alone and this graph shown pasenger information travelling with in titanic
fig=df['number'].value_counts().plot.bar()
fig.set_title('passenger with')
```

```python
#extract numeric part from column
df['trvelling _with_person ']=pd.to_numeric(df['number'],errors='coerce',downcast='integer')
```

```python
#extract categorical part from column
df['treveling_Alone']=np.where(df['trvelling _with_person '].isnull(),df['number'],np.nan)
df.head()
```

```python
df['cabin'].unique()
```

```python
df['Ticket'].unique()
```

```python
df['cabin_num'] = df['cabin'].str.extract('(\d+)') # captures numerical part
df['cabin_cat'] = df['cabin'].str[0] # captures the first letter
df.head()
```

```python
df['cabin_cat'].value_counts().plot(kind='bar')
```

```python
# extract the last bit of ticket as number
df['ticket_num'] = df['Ticket'].apply(lambda s: s.split()[-1])
df['ticket_num'] = pd.to_numeric(df['ticket_num'],
                                   errors='coerce',
                                   downcast='integer')

# extract the first part of ticket as category
df['ticket_cat'] = df['Ticket'].apply(lambda s: s.split()[0])
df['ticket_cat'] = np.where(df['ticket_cat'].str.isdigit(), np.nan,
                              df['ticket_cat'])
df.head(20)
```
